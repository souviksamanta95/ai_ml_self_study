# Text Cleaning and Preprocessing in NLP
This notebook covers essential text cleaning and preprocessing steps for Natural Language Processing (NLP). Each step includes intuitive explanations, practical notes, and best practices.

## Why Text Cleaning?
Raw text data is messy: it contains noise, irrelevant symbols, and inconsistencies. Cleaning and preprocessing text is crucial for building robust NLP models.
- Removes noise (punctuation, numbers, URLs)
- Standardizes text (lowercasing, whitespace normalization)
- Prepares text for tokenization, vectorization, and modeling
- Improves model accuracy and generalization

```python
# 1. Basic Text Cleaning Function
import re
def clean_text(text):
    text = text.lower()  # Lowercase: standardizes text, reduces vocabulary size
    text = re.sub(r'http\S+', '', text)  # Remove URLs: URLs rarely add meaning
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove punctuation & numbers: keeps only letters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces: ensures clean tokenization
    return text
# Example
sample = 'Check out https://github.com! NLP is fun :)'
print(clean_text(sample))
```

### Notes:
- Lowercasing helps treat 'Apple' and 'apple' as the same word.
- Removing URLs, punctuation, and numbers reduces noise.
- Always check if removing numbers/punctuation is appropriate for your task (e.g., sentiment, product codes).

```python
# 2. Tokenization
# Tokenization splits text into words/tokens.
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
tokens = word_tokenize('This is a sample sentence.')
print(tokens)
```

### Notes:
- Tokenization is the foundation for all further text processing.
- NLTK's `word_tokenize` handles punctuation and contractions well.
- For advanced tokenization (e.g., subwords, multi-lingual), use spaCy or Hugging Face tokenizers.

```python
# 3. spaCy Tokenization
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp('This is a sample sentence.')
tokens = [token.text for token in doc]
print(tokens)
```

### Notes:
- spaCy provides fast, robust tokenization and linguistic features (POS, lemma, etc.).
- Use spaCy for production pipelines or when you need more than just splitting text.

```python
# 4. Stopword Removal
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.lower() not in stop_words]
print(filtered)
```

### Notes:
- Stopwords are common words (the, is, and) that add little meaning.
- Removing them can improve model focus, but sometimes they are important (e.g., in sentiment or question answering).
- Always check the stopword list and adapt for your domain/language.

```python
# 5. Stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered]
print(stemmed)
```

### Notes:
- Stemming reduces words to their root form (e.g., 'running' → 'run').
- Stemmers can be aggressive and may produce non-words.
- Use stemming for simple, fast normalization, but prefer lemmatization for better accuracy.

```python
# 6. Lemmatization (spaCy)
lemmatized = [token.lemma_ for token in doc if not token.is_stop]
print(lemmatized)
```

### Notes:
- Lemmatization returns the base form of a word (e.g., 'better' → 'good').
- More accurate than stemming, as it considers context and part of speech.
- Use lemmatization for tasks where word meaning matters (topic modeling, information retrieval).

```python
# 7. Bag-of-Words Vectorization
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,2))  # Unigram + bigram
X = vectorizer.fit_transform(['this is a sample text'])
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### Notes:
- Bag-of-Words (BoW) converts text to a matrix of token counts.
- n-grams (e.g., bigrams) capture short phrases and context.
- BoW ignores word order and meaning, but is simple and effective for many tasks.

```python
# 8. Limiting Vocabulary Size
vectorizer = CountVectorizer(max_features=5000)
corpus = ["This is the first document.", "This document is the second document."]
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
```

### Notes:
- Limiting vocabulary (max_features) helps reduce dimensionality and noise.
- Useful for large corpora or when memory is limited.

```python
# 9. TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(corpus)
print(tfidf.get_feature_names_out())
print(X.toarray())
```

### Notes:
- TF-IDF (Term Frequency-Inverse Document Frequency) weighs words by importance.
- Down-weights common words, up-weights rare but informative words.
- Often outperforms BoW for information retrieval and text classification.

## Final Tips
- Always visualize and inspect your cleaned text and tokens.
- Preprocessing is task-dependent: adapt steps for your domain and language.
- For deep learning, consider using subword tokenization (BPE, WordPiece) and embeddings.
- Save your preprocessing pipeline for reproducibility!
